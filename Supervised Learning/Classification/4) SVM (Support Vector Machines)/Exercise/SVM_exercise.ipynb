```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data, digits.target)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Targets'] = digits.target
df.head(10)
```

```python
groupby = df.groupby('Targets')
groupby.first()
```

```python
groupby.mean()
```

```python
df.head()
```

```python
X = df.drop(['Targets'], axis='columns')
X[:5]
```

```python
y = df.Targets
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
poly_model = SVC(kernel='poly')
```

```python
poly_model.fit(X_train, y_train)
```

```python
y_hat = poly_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Polynomial Model:',accuracy_score(y_test, y_hat))
```

```python
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(X_train, y_train)
```

```python
y_hat = rbf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('RBF Model:',accuracy_score(y_test, y_hat))
```
